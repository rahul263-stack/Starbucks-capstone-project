```python
count_by_agegroup = profile.groupby(pd.cut(profile['age'],np.arange(0, 118+5, 5)))['id'].count()
```

```python
sns.set()
plt.figure(figsize=(12,4))
plt.bar(np.arange(0, 118, 5),count_by_agegroup,width=3.5,align='edge',color = (0.5, 0.6, 0.1,0.6))
plt.xticks(np.arange(0, 118, 5))
plt.xlabel('Age')
plt.ylabel('Persons Counts')
plt.title('The person number of different age')
plt.show()
```
